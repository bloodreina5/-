```python
import numpy as np
import math as m
import scipy.integrate as integr
import matplotlib.pyplot as plt
```

```python
def model(y, t, a,b):
    dy=np.zeros((2,))
    
    r=y[0]
    f=y[1]
    
    dy[0]=2*r-a*r*f+b
    dy[1]=a*r*f-f
   
    
    return dy
```

```python
def paint(r0,f0,a):
    y0=[]
    y0.append(r0)
    y0.append(f0)
    y0=np.array(y0)
    c=a
    b=0
    tmin=0 
    tmax=200
    T=np.arange(tmin,tmax,0.01)
    
    resh=integr.odeint(model, y0, T, args=(c,b))
    
    fig = plt.figure(figsize=(19,5))
    
    ax=fig.add_subplot(1,3,1)
    ax1=fig.add_subplot(1,3,2)
    ax2=fig.add_subplot(1,3,3)
    
    ax.plot(T, resh[:,0])
    ax1.plot(T, resh[:,1])
    ax2.plot(resh[:,0], resh[:,1])
    
    ax.set_xlabel('t', fontsize='15')
    ax.set_ylabel('r', fontsize='15')
    
    ax1.set_xlabel('t', fontsize='15')
    ax1.set_ylabel('f', fontsize='15')
    
    ax2.set_xlabel('r', fontsize='15')
    ax2.set_ylabel('f', fontsize='15')
    
    ax.set_title('r(t)')
    ax1.set_title('f(t)')
    ax2.set_title('фазовый портрет')
    
    ax.grid()
    ax1.grid()
    ax2.grid()
    
    return r0,f0,a
```

```python
print('1 тип фазовых траекторий, r0=f0')
paint(1,1,0.01)
```

```python
print('2 тип фазовых траекторий, r0>>f0(или просто очень большой, портрет примерно похож и при увеличении f0)\nr0=6421(точность до единиц) выстраивается фазовый портрет\nНиже этого значения траектории вели себя не аналитично')
paint(6421,1,0.01)
```

```python
print('3 тип фазовых траекторий, f0>>r0\nf0=5609(точность до единиц) выстраивается фазовый портрет\nНиже этого значения траектории вели себя не аналитично')
paint(1,5609,0.01)
```

```python
def phase_plane_analysys(r0,f0,a):
    y0=[]
    y0.append(r0)
    y0.append(f0)
    y0=np.array(y0)
    tmin=0 
    tmax=200
    T=np.arange(tmin,tmax,0.01)
    b=0
    resh=integr.odeint(model, y0, T, args=(a,b))
    return resh
```

```python
def phase_plane(number):
    fig1=plt.figure(figsize=(10,8))
    for i in range(1,number,1):
        plt.plot(phase_plane_analysys(i,i,0.01)[:,0],phase_plane_analysys(i,i,0.01)[:,1])
    plt.grid()
    plt.title('Множество фазовых траекторий')
    plt.xlabel('r')
    plt.ylabel('f')
    
    print('a=0.01\nФазовых траекторий выведено:')
    return number
```

```python
phase_plane(100)
```

```python
def phase_plane1(number):
    fig1=plt.figure(figsize=(10,8))
    for i in np.arange(0.1,number,0.1):
        plt.plot(phase_plane_analysys(1,1,i)[:,0],phase_plane_analysys(1,1,i)[:,1])
    plt.grid()
    plt.title('Множество фазовых траекторий')
    plt.xlabel('r')
    plt.ylabel('f')
    
    print('перебор альфа с 0.01 до значения:')
    return number
```
